# Using $Tr[² \Delta]$

The formula referred to in the article and from which it was originally taken (DOI: 10.1103/PhysRevA.92.052502) was used for clarity on which part to take the sum of the NOOs. These theorem formulate that the trace of the first order reduced density matrix 1-RDM can be written as the sum of the 1-RDMS's eigenvalues, therefore the trace of the cumulant 2-RDM can also be expressed in terms of the 1-RDM's eigenvalues $n_i$. Then the formula using spin $\sigma$ is (see formula 23 of the reference):
$$ Tr[² \Delta] =  \sum_{i} n_i² - N  $$
Dividiendo dentro de dos porque los Natural Orbital Occupancies se representan para close shell system (los valores van de 0 a 2), por lo que tambien N se debe de dividir dentro de dos para mantenerse en su forma original en el lado derecho de la ecuacion:
$$ Tr[² \Delta] =  \sum_{i} (n_i/2)² - (N/2)  $$
Debido a que se trata de un sistema con dos electrones la ecuacion final queda:
$$ Tr[² \Delta] =  \sum_{i} (n_i)² - 1  $$

```python
#Reading files with the indices of trace
trace_S1pt2_tz = pd.read_csv('./cc-pVTZ/S1_CASPT2/I_trace_CASPT2.csv')
trace_ccsd_tz = pd.read_csv('./cc-pVTZ/CCSD/I_trace_CCSD.csv')
trace_eom_tz = pd.read_csv('./cc-pVTZ/EOM-CCSD/I_trace_EOM.csv')

#Reading the results
Itrace_PT2_tz = trace_S1pt2_tz['I_trace']
Itrace_CCSD_tz = trace_ccsd_tz['I_trace']
Itrace_EOM_tz = trace_eom_tz['I_trace']


#Plotting Ind vrs trace
plt.plot(distance, Ind_EOM_tz, color='crimson', linewidth=1.2, linestyle='dotted', marker='+', markersize=1.5, label='EOM-CCSD $I_{ND}$')
plt.plot(distance, Itrace_EOM_tz, color='crimson', linewidth=1.2, linestyle='solid', marker='+', markersize=1.5, label='EOM-CCSD $Tr[² \Delta]$')
plt.plot(distance, Ind_CCSD_tz, color='blue', linewidth=1.2, linestyle='dotted', marker='o', markersize=1.5, label='CCSD $I_{ND}$')
plt.plot(distance, Itrace_CCSD_tz, color='blue', linewidth=1.2, linestyle='solid', marker='o', markersize=1.5, label='CCSD $Tr[² \Delta]$')
plt.yticks(np.arange(-0.5,0.6,0.1))
plt.xticks(np.arange(0.4,4.2,0.4))
plt.xlabel('H-H distance (Å)')
plt.title('$I_{ND}$ vrs $Tr[² \Delta]$ for EOM-CCSD $S_1$ and CCSD $S_0$')

# Displaying legend
plt.legend()

# SAVING THE PLOT
#plt.savefig('I_S1_basis.png', dpi=(250), bbox_inches='tight')

# Show the plot
plt.show()
```

As can be observed for the case of CCSD, using the formula $Tr[² \Delta]$, a value between -0.4 and -0.5 is reached at a distance of 2.4 Å, until reach a minium of -0.5 at a distance of 3.2 Å, thus having the same behavior but opposite in sign to Matito's formula for non-dynamic correlation. The same happens with EOM-CCSD, however, as it does not show variation in NOOs along the dissociation, the values remain constant.

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

from matplotlib import cm
from matplotlib.ticker import LinearLocator
```

# Potential energy surfaces between $S_0$ and $S_1$

NOTE: The energy for the $S_1$ was taken from XMULTI

```python
df_PES_S0 = pd.read_csv('./S0_CASPT2/indices/S0_CASSCF_PES.csv')
df_PES_S1 = pd.read_csv('./S1_CASPT2/XMULTI/indices/S1_CASSCF_XMULT_PES.csv')
df_PES_S2 = pd.read_csv('./S1_CASPT2/XMULTI/indices/S2_CASSCF_XMULT_PES.csv')

angles = df_PES_S0['distance/angle']
stretch_values = [1.262 + i * 0.05 for i in range(14)]

# Create a 2D grid for X and Y
X, Y = np.meshgrid(stretch_values, angles)

# Initialize a 2D array for Z
Z1 = np.zeros_like(X)
Z2 = np.zeros_like(X)
Z3 = np.zeros_like(X)

# Iteration over each stretching value in the file
for i, stretch in enumerate(stretch_values):
    energy_values_S0 = df_PES_S0[f'{stretch:.3f}']
    energy_S0 = energy_values_S0 * 27.2114
    energy_values_S1 = df_PES_S1[f'{stretch:.3f}']
    energy_S1 = energy_values_S1 * 627.5
    energy_values_S2 = df_PES_S2[f'{stretch:.3f}']
    #dif = energy_values_S1 - energy_values_S0
    #dif_conver = dif *  27.2114
    #print(dif)
    Z1[:, i] = energy_values_S0
    Z2[:, i] = energy_values_S1
    Z3[:, i] = energy_values_S2
    #Z1[:, i] = energy_S0
    #Z2[:, i] = energy_S1

#dif_conver = dif *  27.2114
#print(dif_conver)

# Create a new figure and 3D axes
fig = plt.figure(figsize=(14, 8))
ax = fig.add_subplot(111, projection='3d')

surf = ax.plot_surface(X, Y, Z1, cmap=cm.coolwarm, linewidth=0.5, edgecolor='k')
surf = ax.plot_surface(X, Y, Z2, cmap=cm.coolwarm, linewidth=0, edgecolor='k')
#surf = ax.plot_surface(X, Y, Z3, cmap=cm.coolwarm, linewidth=0, edgecolor='k')
ax.azim = -35
#ax.dist = 10
ax.elev = 10

ax.set_xlabel('Stretching distance (Å)')
ax.set_ylabel('Twist angle')
ax.set_zlabel('Energy in A.U.')
plt.title('Fulvene PES using CASSCF(6,6) cc-pVDZ: Stretching vrs twist angle')

fig.colorbar(surf, label='Energy')

#plt.savefig('PES_CCSD_torsion.png', dpi=(250), bbox_inches='tight')

plt.show()
```

# Correlation indices: dynamic and non-dynamic

## Ground state $S_0$

### Non-dynamic indices for $S_0$

```python
data_list1 = []

# Iterar sobre los nombres de archivos esperados
stretch_values = [1.262 + i * 0.05 for i in range(11)]
for stretch in stretch_values:
    filename1 = f"./S0_CASPT2/indices/casindices_stretch.{stretch:.3f}.csv"
    
    # Cargar el archivo CSV en un DataFrame
    df = pd.read_csv(filename1)
    
    # Agregar los datos relevantes a la lista
    data_list1.append({
        'stretch': stretch,
        'angles': df['angle'].values,
        'It_values': df['Ind'].values
    })

# Preparar datos para el diagrama de calor
angles = data_list1[0]['angles']  # Tomar los ángulos del primer archivo (son los mismos para todos)
num_angles = len(angles)  # Número de ángulos por archivo
num_stretch_values = len(stretch_values)  # Número de valores de stretching

# Crear matriz para almacenar los valores de 'It' como matriz de calor
It_matrix1 = np.zeros((num_angles, num_stretch_values))

# Llenar la matriz con los valores de 'It' correspondientes
for i, data in enumerate(data_list1):
    stretch_idx = i
    It_values1 = data['It_values']
    It_matrix1[:, stretch_idx] = It_values1


#####################################
# Plotting for the second graph 
data_list2 = []

for stretch in stretch_values:
    filename2 = f"./S0_CASPT2/indices/pt2indices_stretch.{stretch:.3f}.csv"
    
    # Cargar el archivo CSV en un DataFrame
    df = pd.read_csv(filename2)
    
    # Agregar los datos relevantes a la lista
    data_list2.append({
        'stretch': stretch,
        'angles': df['angle'].values,
        'It_values': df['Ind'].values
    })


# Crear matriz para almacenar los valores de 'It' como matriz de calor
It_matrix2 = np.zeros((num_angles, num_stretch_values))

# Llenar la matriz con los valores de 'It' correspondientes
for i, data in enumerate(data_list2):
    stretch_idx = i
    It_values2 = data['It_values']
    It_matrix2[:, stretch_idx] = It_values2



########################################
#Data for the third plot
data_list3 = []
for stretch in stretch_values:
    filename3 = f"./CCSD/indices/indices_stretch.{stretch:.3f}.csv"
    
    # Cargar el archivo CSV en un DataFrame
    df = pd.read_csv(filename3)
    
    # Agregar los datos relevantes a la lista
    data_list3.append({
        'stretch': stretch,
        'angles': df['angle'].values,
        'It_values': df['Ind'].values
    })


# Crear matriz para almacenar los valores de 'It' como matriz de calor
It_matrix3 = np.zeros((num_angles, num_stretch_values))

# Llenar la matriz con los valores de 'It' correspondientes
for i, data in enumerate(data_list3):
    stretch_idx = i
    It_values3 = data['It_values']
    It_matrix3[:, stretch_idx] = It_values3



#######################3###############
#Ploting the two graphis together 
fig, axs = plt.subplots(1, 3, figsize=(12, 5), layout='constrained')
#First heatmap

vmax = 0.085
vmin = 0.01

im1 = axs[0].imshow(It_matrix1, aspect='auto', cmap='bwr', interpolation='nearest')
axs[0].set_title('CASSCF(2,2)', fontsize=18)
axs[0].set_xlabel('Stretching', fontsize=12)
axs[0].set_ylabel('Twist angle', fontsize=13)
axs[0].set_xticks(np.arange(num_stretch_values), [f'{stretch:.3f}' for stretch in stretch_values], rotation=45)
axs[0].set_yticks(np.arange(num_angles), angles)
axs[0].set_yticklabels(angles)
fig.colorbar(im1, ax=axs[0], label='$I_{ND}$')

# Segundo heatmap
im2 = axs[1].imshow(It_matrix2, aspect='auto', cmap='bwr',   interpolation='nearest')
axs[1].set_title('CASPT2(2,2)', fontsize=18)
axs[1].set_xlabel('Stretching', fontsize=12)
#axs[1].set_ylabel('Bending angle')
axs[1].set_xticks(np.arange(num_stretch_values), [f'{stretch:.3f}' for stretch in stretch_values], rotation=45)
axs[1].set_yticks([])  # Quitar los ticks del eje y
axs[1].set_yticklabels([])
fig.colorbar(im2, ax=axs[1], label='$I_{ND}$')

#Third heatmap

im3 = axs[2].imshow(It_matrix3, aspect='auto', cmap='bwr',  interpolation='nearest')
axs[2].set_title('CCSD',  fontsize=18)
axs[2].set_xlabel('Stretching', fontsize=12)
#axs[2].set_ylabel('Bending angle')
axs[2].set_xticks(np.arange(num_stretch_values), [f'{stretch:.3f}' for stretch in stretch_values], rotation=45)
axs[2].set_yticks([])  # Quitar los ticks del eje y
axs[2].set_yticklabels([])  # Quitar las etiquetas del eje y
fig.colorbar(im3, ax=axs[2], label='$I_{ND}$')
#cbar = fig.colorbar(im2, location='bottom', label='Non-dynamic indices ($I_{ND}$)', shrink=1)
#tick_font_size = 12
#cbar.ax.tick_params(labelsize=tick_font_size)


#plt.savefig('Ind_torsion_S0.png', dpi=(250), bbox_inches='tight')

plt.show()
```

```python
data_list = []

# Iterar sobre los nombres de archivos i

stretch_values = [1.262 + i * 0.05 for i in range(14)]
for stretch in stretch_values:
    filename = f"./CCSD/indices/corr_stretch.{stretch:.3f}.csv"
    
    # Cargar el archivo CSV en un DataFrame
    df = pd.read_csv(filename)
    
    # Agregar los datos relevantes a la lista
    data_list.append({
        'stretch': stretch,
        'angles': df['angle'].values,
        'It_values': df['corr_percentage'].values
    })

# Preparar datos para el diagrama de calor
angles = data_list[0]['angles']  # Tomar los ángulos del primer archivo (son los mismos para todos)
num_angles = len(angles)  # Número de ángulos por archivo
num_stretch_values = len(stretch_values)  # Número de valores de stretching

# Crear matriz para almacenar los valores de 'It' como matriz de calor
It_matrix = np.zeros(( num_angles, num_stretch_values))

# Llenar la matriz con los valores de 'It' correspondientes
for i, data in enumerate(data_list):
    stretch_idx = i
    It_values = data['It_values']
    It_matrix[:, stretch_idx] = It_values

vmax = 0.1050
vmin = 0.050


# Crear el diagrama de calor
plt.figure(figsize=(10,6))
plt.imshow(It_matrix, aspect='auto',  cmap='bwr', interpolation='nearest')
#plt.imshow(It_matrix, aspect='auto',  cmap='bwr', interpolation='nearest')

# Configurar etiquetas de los ejes
plt.xlabel('Stretching')
plt.ylabel('Twist angle')
plt.title('Energy difference CCSD-HF')

# Definir los valores reales de los ejes
plt.yticks(np.arange(num_angles), angles)
plt.xticks(np.arange(num_stretch_values), [f'{stretch:.3f}' for stretch in stretch_values], rotation=45)


# Añadir barra de color
plt.colorbar(label='Energy CCSD-HF a.u.')
#plt.savefig('Ind_CCSD_D2d_S1.png', dpi=(250), bbox_inches='tight')

# Mostrar el diagrama de calor
plt.show()
```

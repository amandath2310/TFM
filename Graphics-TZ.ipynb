## For the case of $S_0$

```python
#Reading the files
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt

#Reading files for CASSCF
S0cas_path_dz = './cc-pVDZ/S0_CASSCF/Indices_CASSCF.csv'
S0cas_path_tz = './cc-pVTZ/S0_CASSCF/Indices_CASSCF.csv'
df_S0cas_dz = pd.read_csv(S0cas_path_dz)
df_S0cas_tz = pd.read_csv(S0cas_path_tz)


#Reading files for CASPT2
S0pt2_path_dz = './cc-pVDZ/S0_CASPT2/Indices_CASPT2.csv'
S0pt2_path_tz = './cc-pVTZ/S0_CASPT2/Indices_CASPT2.csv'
df_S0pt2_dz = pd.read_csv(S0pt2_path_dz)
df_S0pt2_tz = pd.read_csv(S0pt2_path_tz)


#Reading files for CCSD
ccsd_path_dz = './CCSD/Indices_CCSD.csv'
ccsd_path_tz = './cc-pVTZ/CCSD/Indices_CCSD.csv'
df_ccsd_dz = pd.read_csv(ccsd_path_dz)
df_ccsd_tz = pd.read_csv(ccsd_path_tz)

#Extracting the indices from the files
#For CASSCF
distance = df_S0cas_dz['Distance']
Id_CAS_dz = df_S0cas_dz['Id']
Ind_CAS_dz = df_S0cas_dz['Ind']
It_CAS_dz = df_S0cas_dz['It']
Id_CAS_tz = df_S0cas_tz['Id']
Ind_CAS_tz = df_S0cas_tz['Ind']
It_CAS_tz = df_S0cas_tz['It']

#For CASPT2
Id_PT2_dz = df_S0pt2_dz['Id']
Ind_PT2_dz = df_S0pt2_dz['Ind']
It_PT2_dz = df_S0pt2_dz['It']
Id_PT2_tz = df_S0pt2_tz['Id']
Ind_PT2_tz = df_S0pt2_tz['Ind']
It_PT2_tz = df_S0pt2_tz['It']

#For CCSD
Id_CCSD_dz = df_ccsd_dz['Id']
Ind_CCSD_dz = df_ccsd_dz['Ind']
It_CCSD_dz = df_ccsd_dz['It']
Id_CCSD_tz = df_ccsd_tz['Id']
Ind_CCSD_tz = df_ccsd_tz['Ind']
It_CCSD_tz = df_ccsd_tz['It']



#Plotting the indices
plt.plot(distance, Ind_CAS_tz, color='orange', linewidth=1.2, linestyle='dotted')
plt.plot(distance, Ind_PT2_tz, color='royalblue', linewidth=1.2, linestyle='dotted')
plt.plot(distance, It_CAS_tz, color='orange', linewidth=1.2, linestyle='dashed')
plt.plot(distance, It_PT2_tz, color='royalblue', linewidth=1.2, linestyle='dashed')
plt.plot(distance, Id_CAS_tz, color='orange', linewidth=1.2, linestyle='solid', label='CASSCF cc-pVTZ')
plt.plot(distance, Id_PT2_tz, color='royalblue', linewidth=1.2, linestyle='solid', label='CASPT2 cc-pVTZ')
plt.plot(distance, Ind_PT2_dz, color='black', linewidth=1.2, linestyle='dotted')
plt.plot(distance, It_PT2_dz, color='black', linewidth=1.2, linestyle='dashed')
plt.plot(distance, Id_PT2_dz, color='black', linewidth=1.2, linestyle='solid', label='CASPT2 cc-pVDZ')
plt.xlabel('H-H distance (Å)')

# Displaying legend
plt.legend()

# SAVING THE PLOT
plt.savefig('I_CAS-PT2_S0.png', dpi=(250), bbox_inches='tight')

# Show the plot
plt.show()


#Plotting the indices
plt.plot(distance, Ind_CCSD_tz, color='crimson', linewidth=1.2, linestyle='dotted')
plt.plot(distance, It_CCSD_tz, color='crimson', linewidth=1.2, linestyle='dashed')
plt.plot(distance, Id_CCSD_tz, color='crimson', linewidth=1.2, linestyle='solid', label='cc-pVTZ')
plt.plot(distance, Ind_CCSD_dz, color='blue', linewidth=1.2, linestyle='dotted')
plt.plot(distance, It_CCSD_dz, color='blue', linewidth=1.2, linestyle='dashed')
plt.plot(distance, Id_CCSD_dz, color='blue', linewidth=1.2, linestyle='solid', label='cc-pVDZ')
plt.xlabel('H-H distance (Å)')

# Displaying legend
plt.legend()

# SAVING THE PLOT
plt.savefig('I_CCSD_basis.png', dpi=(250), bbox_inches='tight')

# Show the plot
plt.show()
```
